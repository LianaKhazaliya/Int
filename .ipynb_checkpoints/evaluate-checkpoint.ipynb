```python
import csv
import numpy
import pandas as pd
from sklearn.decomposition import PCA
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
import re
import os, sys
from math import sqrt
```

```python
def get_data_about_smb(input_file, names, *parameters):
    data = []
    for name in names:
        with open(input_file, mode = 'r') as csvfile:
            reader = csv.DictReader(csvfile) 
            for row in reader:
                if (row['Mark'] == name):
                    d = {'Mark' : name}
                    for par in parameters:
                        if par[1] == 'int':
                            d[par[0]] = int(row[par[0]].replace(' ', ''))
                        else:
                            d[par[0]] = str(row[par[0]])
                    data.append(d)

    return data
```

```python
data = pd.DataFrame(get_data_about_smb('DATA1.csv', ['A4'], ['Run', 'int'], ['Year', 'int'], ['Cost', 'int']))
```

```python
def coefficient(d, year, run):    
    d_y = d['Year']
    d_r = d['Run']
    C = 1
    
    C_1 = (1/(max(1, abs(d['Year']-year)))**(1/4))
    if d_y < year:
        C *= C_1
    else:
        C /= C_1
        
    C_2 = (5000/(max(abs(run - d_r), 5000)))**(1/10)
    
    if d_r > run:
        C *= C_2
    else:
        C /= C_2

    return C

C = 0
cnt = 0
for x in range(len(data.index)):

    d = data.loc[x]
    print(d)
    c = coefficient(d, 1996, 218000)
    print(1/c)
    if abs(1/c-1) < 0.4:
        C += d['Cost']/c
        cnt += 1

print(C/cnt)
    
```

```python
def get_cost(mark, year, run):
    data = pd.DataFrame(get_data_about_smb('DATA1.csv', [mark], ['Run', 'int'], ['Year', 'int'], ['Cost', 'int']))
    print(data)
    C = 0
    cnt = 0
    for x in range(len(data.index)):

        d = data.loc[x]
        print(d)
        c = coefficient(d, year, run)
        print(c)
        if abs(c-1) < 0.1:
            C += d['Cost']/c
            cnt += 1

    print(round(C/cnt, 2))
```

```python
def get_data_about_smb2(input_file, names, *parameters):
    data = []
    for name in names:
        with open(input_file, mode = 'r') as csvfile:
            reader = csv.DictReader(csvfile) 
            for row in reader:
                if (row['Title'] == name):
                    d = {'Title' : name}
                    for par in parameters:
                        if par[1] == 'int':
                            d[par[0]] = int(row[par[0]].replace(' ', ''))
                        else:
                            d[par[0]] = str(row[par[0]])
                    data.append(d)

    return data
```

```python
def get_cost2(title, year, run):
    data = pd.DataFrame(get_data_about_smb2('DATA1.csv', [title], ['Run', 'int'], ['Year', 'int'], ['Cost', 'int']))
    print(data)
    C = 0
    cnt = 0
    for x in range(len(data.index)):

        d = data.loc[x]
        print(d)
        c = coefficient(d, year, run)
        print(c)
        if abs(c-1) < 0.1:
            C += d['Cost']/c
            cnt += 1

    print(round(C/cnt, 2))
```

```python
get_cost2('Audi A4 B9', 2014, 30000)
```
